# Encapsulation

```python
class Student:
    def __init__(self, name, marks, roll_no):
        self.name = name
        self.marks = marks
        self.roll_no = roll_no
        print("Student class has been called")
    def is_distinction(self):
        if self.marks > 75:
            print("Students got distinction")
        else:
            print("Students didn't get distinction")
```

```python
s1 = Student("Mayank", 22, 12)
```

```python
s1.is_distinction()
```

```python
s1.marks = 76
```

```python
s1.is_distinction()
```

## Access modifier

Its not a good practice to have data properties of our object/instance to be open for access.



```python
class Student_better:
    def __init__(self, name, marks, roll_no):
        self.__name = name
        self.__marks = marks
        self.__roll_no = roll_no
        print("Student class has been called")

    # getter
    def get_marks(self, passcode):
        if (passcode == "1234"):
            print("The marks are "+ str(self.__marks))
        else:
            print("access denied")
    # setter
    def set_marks(self, new_score, teacherid):
        if (teacherid[0] == "0001"):
            self.__marks = new_score
        else:
            print("You are not registered with us and not allowed to change the student marks")
```

```python
s2 = Student_better("Mayank", 22, 12)
```

```python
s2.get_marks()
```

```python
s2.marks()
```

```python
s2.set_marks(60)
```

```python
s2.get_marks()
```
